```python
import random

risultati = []
for _ in range(100):
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    somma = dado1 + dado2
    risultati.append(somma)

# Conta le frequenze
frequenze = {}
for valore in risultati:
    frequenze[valore] = frequenze.get(valore, 0) + 1

# Trova la somma più frequente
somma_frequente = max(frequenze, key=frequenze.get)
conteggio = frequenze[somma_frequente]

print("La somma più frequente è:", somma_frequente)
print("Appare", conteggio, "volte")

print(somma_frequente/100)
```
