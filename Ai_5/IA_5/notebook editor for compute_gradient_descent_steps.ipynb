```python
import dataiku
import pandas as pd
import numpy as np

# =====================
# READ DATA
# =====================

df = dataiku.Dataset("regression_training").get_dataframe()

x = df["x"].values
y = df["y"].values

# =====================
# FUNCTIONS
# =====================

def compute_cost(x, y, w, b):
    m = len(x)
    cost = 0
    for i in range(m):
        cost += (w*x[i] + b - y[i])**2
    return cost/(2*m)


def compute_gradient(x, y, w, b):
    m = len(x)
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        err = (w*x[i] + b - y[i])
        dj_dw += err * x[i]
        dj_db += err

    return dj_dw/m, dj_db/m


# =====================
# GRADIENT DESCENT
# =====================

w = 0
b = 0
alpha = 0.02
iterations = 40

rows = []

for i in range(iterations):

    cost = compute_cost(x,y,w,b)
    dj_dw, dj_db = compute_gradient(x,y,w,b)

    rows.append({
        "iteration": i,
        "w": w,
        "b": b,
        "cost": cost,
        "dj_dw": dj_dw,
        "dj_db": dj_db
    })

    w = w - alpha*dj_dw
    b = b - alpha*dj_db


# =====================
# SAVE OUTPUT
# =====================

out_df = pd.DataFrame(rows)

dataiku.Dataset("gradient_descent_steps").write_with_schema(out_df)
```
