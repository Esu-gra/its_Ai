```python

import pandas as pd
import psycopg
import seaborn as sns
import matplotlib.pyplot as plt

conn = psycopg.connect(
    host='its_postgresql',
    port=5432,
    dbname='techstore_DW',
    user='postgres',
    password='postgres'
)

print('Connessione OK')
```

```python
# Query al database
df = pd.read_sql(
    '''
    SELECT 
        categoria, 
        COUNT(*) AS n
    FROM fact_vendite fv
    JOIN dim_prodotto dp 
        ON fv.prodotto_sk = dp.prodotto_sk
    GROUP BY categoria
    ''',
    conn
)

# Calcolo della probabilit√† (frequenza relativa)
df['prob'] = df['n'] / df['n'].sum()

# Mostra il risultato
df
```

```python
df_eta = pd.read_sql(
    '''
    SELECT 
        fascia_eta,
        COUNT(*) AS n
    FROM fact_vendite fv
    JOIN dim_cliente dc 
        ON fv.cliente_sk = dc.cliente_sk
    GROUP BY fascia_eta
    ORDER BY fascia_eta
    ''',
    conn
)

# Calcolo della distribuzione (frequenza relativa)
df_eta['prob'] = df_eta['n'] / df_eta['n'].sum()

df_eta
```

```python
df_ct = pd.read_sql(
    '''
    SELECT 
        fascia_eta,
        categoria,
        COUNT(*) AS n
    FROM fact_vendite fv
    JOIN dim_cliente dc 
        ON fv.cliente_sk = dc.cliente_sk
    JOIN dim_prodotto dp 
        ON fv.prodotto_sk = dp.prodotto_sk
    GROUP BY fascia_eta, categoria
    ORDER BY fascia_eta, categoria
    ''',
    conn
)

# Creazione tabella pivot
pivot = df_ct.pivot(
    index='fascia_eta',
    columns='categoria',
    values='n'
)

# Heatmap della pivot
sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')








```

```python
#esercizio 4

# Seleziono la riga
riga_1825 = pivot.loc['18-24']

# Calcolo la probabilit√† condizionata:
p_laptop_given_1825 = riga_1825['Laptop'] / riga_1825.sum()

p_laptop_given_1825
```

```python
#esercizio 5
# Sommo i valori per la colonna Smartphone (totale smartphone venduti)
totale_smartphone = pivot['Smartphone'].sum()

# Prendo la cella corrispondente a (18‚Äì24, Smartphone)
smartphone_1824 = pivot.loc['18-24', 'Smartphone']

# Probabilit√† condizionata P(18-24 | Smartphone)
p_1824_dato_smartphone = smartphone_1824 / totale_smartphone
print(round(p_1824_dato_smartphone,4))
```

```python
import pandas as pd

import psycopg
import seaborn as sns
import matplotlib.pyplot as plt

conn = psycopg.connect(
    host='its_postgresql',
    port=5432,
    dbname='techstore_DW',
    user='postgres',
    password='postgres'
)

df = pd.read_sql(
    '''
    SELECT 
        dc.fascia_eta,
        dp.categoria,
        fv.prezzo_unitario,
        COUNT(*) AS n
    FROM fact_vendite fv
    JOIN dim_cliente dc ON fv.cliente_sk = dc.cliente_sk
    JOIN dim_prodotto dp ON fv.prodotto_sk = dp.prodotto_sk
    GROUP BY 1,2,3
    ''',
    conn
)
#Aggiungo la fascia costo usando prezzo_unitario
df['fascia_costo'] = pd.cut(
    df['prezzo_unitario'],
    bins=[0, 300, 800, 20000],
    labels=['basso', 'medio', 'alto']
)
#Filtra 18‚Äì24 + costo alto
filtro = df[
    (df['fascia_eta'] == '18-24') &
    (df['fascia_costo'] == 'alto')
]

#Numeratore: solo categoria ‚ÄúLaptop‚Äù
num_laptop = filtro[filtro['categoria'] == 'Laptop']['n'].sum()

#Denominatore: tutte le categorie in 18‚Äì24 costo alto
den = filtro['n'].sum()

#probabilit√† condizionata
p_laptop_18_24_costo_alto = num_laptop / den
print(round(p_laptop_18_24_costo_alto,4))



```
