# Demo: Regressione Lineare (Statistica → Machine Learning)
Questo notebook è pensato per accompagnare la lezione: 
- Parte A: regressione lineare semplice (univariata) su un esempio “case”
- Parte B: train/test split e metriche (MAE, RMSE, R²)
- Parte C: diagnostica residui
- Parte D: regressione multivariata e regolarizzazione (Ridge/Lasso)

**Nota:** esegui le celle in ordine.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

np.random.seed(42)
```

## Parte A — Regressione lineare semplice (univariata)
Obiettivo: prevedere il **prezzo** di una casa in funzione della **superficie**.

Modello: $\hat{y} = \beta_0 + \beta_1 x$

```python
# Dataset didattico (perfettamente lineare) — prezzi in Euro
X = np.array([[50], [75], [100], [125], [150]])
y = np.array([150000, 225000, 300000, 375000, 450000])

df = pd.DataFrame({'superficie_m2': X.ravel(), 'prezzo_eur': y})
df
```

```python
# Fit del modello OLS (scikit-learn)
model = LinearRegression()
model.fit(X, y)

beta0 = model.intercept_
beta1 = model.coef_[0]
print(f"β0 (intercetta) = {beta0:,.2f} €")
print(f"β1 (pendenza)   = {beta1:,.2f} € per m²")

# Previsione esempio: 110 m²
pred_110 = model.predict([[110]])[0]
print(f"Prezzo stimato per 110 m²: {pred_110:,.2f} €")
```

```python
# Visualizzazione: punti e retta
x_plot = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
y_plot = model.predict(x_plot)

plt.figure(figsize=(7,4))
plt.scatter(X, y, label="Dati")
plt.plot(x_plot, y_plot, label="Retta stimata")
plt.xlabel("Superficie (m²)")
plt.ylabel("Prezzo (€)")
plt.title("Regressione lineare semplice")
plt.legend()
plt.show()
```

## Parte B — Dati più realistici, train/test split e metriche
Ora creiamo un dataset più “sporco” con rumore: il modello non sarà perfetto.

Valuteremo MAE, RMSE e R² su train e su test.

```python
df = pd.read_csv("immobili.csv")
#print("Colonne trovate:", df.columns.tolist())
```

```python
# usiamo dati reali presi di 100 immobili da immobiliare.it relativi ad un quartiere di Roma
df = pd.read_csv("immobili.csv", sep=";")

Xn = df[["superficie_m2"]].values   # feature (2D)
yn = df["prezzo_eur"].values        # target (1D)

# Split
X_train, X_test, y_train, y_test = train_test_split(Xn, yn, test_size=0.25, random_state=42)

lin = LinearRegression()
lin.fit(X_train, y_train)

def eval_metrics(y_true, y_pred, label=""):
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"{label:>6}  MAE={mae:,.0f}   R²={r2:.3f}")

eval_metrics(y_train, lin.predict(X_train), "train")
eval_metrics(y_test,  lin.predict(X_test),  "test")
```

```python
# Visualizzazione: dati + linea stimata
x_plot = np.linspace(Xn.min(), Xn.max(), 200).reshape(-1,1)
y_plot = lin.predict(x_plot)

plt.figure(figsize=(7,4))
plt.scatter(X_train, y_train, alpha=0.6, label="Train")
plt.scatter(X_test, y_test, alpha=0.6, label="Test")
plt.plot(x_plot, y_plot, label="Retta stimata")
plt.xlabel("Superficie (m²)")
plt.ylabel("Prezzo (€)")
plt.title("Dataset rumoroso + regressione lineare")
plt.legend()
plt.show()
```

## Parte C — Residui e diagnostica
Residuo: $e_i = y_i - \hat{y}_i$. 

Guardiamo:
- residui vs valori predetti (pattern? ventaglio?)
- distribuzione dei residui

```python
# Residui su test
y_pred_test = lin.predict(X_test)
resid = y_test - y_pred_test

plt.figure(figsize=(6,4))
plt.scatter(y_pred_test, resid, alpha=0.7)
plt.axhline(0, linewidth=1)
plt.xlabel("Valori predetti (test)")
plt.ylabel("Residui (y - ŷ)")
plt.title("Residui vs fitted (test)")
plt.show()

plt.figure(figsize=(6,4))
plt.hist(resid, bins=20)
plt.xlabel("Residuo")
plt.ylabel("Frequenza")
plt.title("Distribuzione dei residui (test)")
plt.show()
```

## Parte D — Regressione multivariata + regolarizzazione
Esempio: dataset **diabetes** (scikit-learn) con più feature.

Confrontiamo:
- LinearRegression
- Ridge (L2)
- Lasso (L1)

Useremo una **Pipeline** con StandardScaler per evitare leakage.

```python
from sklearn.datasets import load_diabetes

data = load_diabetes(as_frame=True)
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

pip_lin = Pipeline([('scaler', StandardScaler()), ('model', LinearRegression())])
pip_ridge = Pipeline([('scaler', StandardScaler()), ('model', Ridge(alpha=1.0))])
pip_lasso = Pipeline([('scaler', StandardScaler()), ('model', Lasso(alpha=0.05, max_iter=10000))])

for name, pip in [('Linear', pip_lin), ('Ridge', pip_ridge), ('Lasso', pip_lasso)]:
    pip.fit(X_train, y_train)
    y_pred = pip.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    print(f"{name:6}  MAE={mae:.2f}  RMSE={rmse:.2f}  R²={r2:.3f}")
```

```python
# Cross-validation (R²) per scegliere alpha in modo più robusto
alphas = [0.01, 0.1, 1, 10, 100]
mean_scores = []
std_scores = []

for a in alphas:
    pip = Pipeline([('scaler', StandardScaler()), ('model', Ridge(alpha=a))])
    scores = cross_val_score(pip, X, y, cv=5, scoring='r2')
    mean_scores.append(scores.mean())
    std_scores.append(scores.std())

plt.figure(figsize=(7,4))
plt.plot([str(a) for a in alphas], mean_scores, marker='o')
plt.xlabel("alpha (Ridge)")
plt.ylabel("R² medio (CV=5)")
plt.title("Selezione alpha con cross-validation")
plt.show()

best_a = alphas[int(np.argmax(mean_scores))]
print("Miglior alpha (Ridge) secondo CV:", best_a)
```

```python
# Coefficienti: interpretazione (su dati standardizzati)
best_ridge = Pipeline([('scaler', StandardScaler()), ('model', Ridge(alpha=best_a))])
best_ridge.fit(X_train, y_train)

coefs = best_ridge.named_steps['model'].coef_
coef_series = pd.Series(coefs, index=X.columns).sort_values()

plt.figure(figsize=(7,4))
plt.barh(coef_series.index, coef_series.values)
plt.xlabel("Coefficiente (standardizzato)")
plt.title("Importanza / direzione delle feature (Ridge)")
plt.show()

coef_series.tail(5)
```

## (Extra) Gradient descent “da zero” su univariata
Qui implementiamo una versione minimale per vedere J che scende.
Non è necessario per usare scikit-learn, ma aiuta l’intuizione.

```python
# Univariata: usiamo il dataset rumoroso (Xn, yn) e stimiamo w,b con gradient descent
Xgd = Xn.ravel()
ygd = yn

w = 0.0
b = 0.0
alpha = 1e-6
m = len(Xgd)

def compute_cost(w, b):
    yhat = w*Xgd + b
    return np.sum((yhat - ygd)**2) / (2*m)

cost_hist = []
for it in range(3000):
    yhat = w*Xgd + b
    dj_dw = np.sum((yhat - ygd)*Xgd) / m
    dj_db = np.sum((yhat - ygd)) / m
    w -= alpha * dj_dw
    b -= alpha * dj_db
    if it % 50 == 0:
        cost_hist.append(compute_cost(w,b))

print("w, b:", w, b)
print("Costo finale:", compute_cost(w,b))
```

```python
plt.figure(figsize=(7,4))
plt.plot(np.arange(len(cost_hist))*50, cost_hist)
plt.xlabel("Iterazioni")
plt.ylabel("Costo J")
plt.title("Discesa del gradiente: costo nel tempo")
plt.show()
```
